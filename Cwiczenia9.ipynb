### Naiwny Algorytm Bayesa

Naiwny algorytm Bayesa (ang. Naive Bayes) jest klasycznym algorytmem probabilistycznym wykorzystywanym głównie w klasyfikacji. Jest to model oparty na twierdzeniu Bayesa z założeniem o niezależności cech. Oznacza to, że dla danego klasyfikowanego obiektu, poszczególne cechy (zmienne) traktowane są jako niezależne, co znacznie upraszcza obliczenia.

## Twierdzenie Bayesa

Twierdzenie Bayesa opisuje zależność między warunkową a priori i posteriori prawdopodobieństwa:

$$
P(C \mid X) = \frac{P(X \mid C) P(C)}{P(X)}
$$

gdzie:

- $ P(C \mid X) $ - prawdopodobieństwo, że obiekt należy do klasy $ C $, biorąc pod uwagę cechy $ X $ (posterior),
- $ P(X \mid C) $ - prawdopodobieństwo wystąpienia cech $ X $, jeśli obiekt należy do klasy $ C $ (likelihood),
- $ P(C) $ - prawdopodobieństwo priori, że obiekt należy do klasy $ C $,
- $ P(X) $ - prawdopodobieństwo wystąpienia cech $ X $ (często traktowane jako stała, ignorowane w procesie klasyfikacji).

<br>

## Założenia Naiwnego Bayesa

Naiwny Bayes zakłada, że cechy są niezależne. Oznacza to, że w obliczeniach traktuje się każdą cechę oddzielnie, ignorując wzajemne zależności między nimi. Dzięki temu algorytm jest bardzo szybki i efektywny, choć nie zawsze osiąga najlepsze wyniki w porównaniu z bardziej zaawansowanymi modelami.

## Modelowanie w Naiwnym Bayesie

Aby obliczyć prawdopodobieństwo, że obiekt należy do klasy $ C $, stosujemy następujący wzór:

$$
P(C \mid X) = P(C) \prod_{i=1}^{n} P(x_i \mid C) \frac{1}{P(X)}
$$

$$
\prod_{i=1}^{n} P(x_i \mid C) = P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C)
$$

$$
P(C \mid X) = \frac {P(C) \cdot P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C) }{P(X)}
$$

<br>

Możemy też pominąć mianownik, bo jest taki sam dla wszystkich klas i nie wpływa na klasyfikację, wtedy:

$$
P(C \mid X) \propto P(C) \cdot P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C)
$$

gdzie:

- $ x_1, x_2, \dots, x_n $ to cechy obiektu $ X $,
- $ P(x_i \mid C) $ to prawdopodobieństwo wystąpienia cechy $ x_i $, gdy obiekt należy do klasy $ C $.

W praktyce algorytm oblicza prawdopodobieństwa dla każdej klasy i wybiera tę, która daje największe prawdopodobieństwo.

### Krok po kroku:

1. Oblicz priorytetowe prawdopodobieństwo $ P(C) $ dla każdej klasy $ C $, czyli prawdopodobieństwo, że dany obiekt należy do klasy $ C $.

$$
P(C) = \frac{\text{liczba próbek klasy } C}{\text{łączna liczba próbek}}
$$

2. Oblicz prawdopodobieństwa warunkowe $ P(x_i \mid C) $ dla każdej cechy $ x_i $ i klasy $ C $. Można to zrobić na podstawie częstości występowania cechy $ x_i $ w klasie $ C $.

3. Zastosuj wzór Bayesa dla każdej klasy $ C $ i wybierz tę klasę, dla której $ P(C \mid X) $ jest największe.

Zadanie polega na wyliczeniu, do której klasy będzie należał nowy obiekt.


### Obliczanie prawdopodobieństw przy użyciu wzoru Bayesa

## 1. Oblicz prawdopodobieństwa priorytetowe $ P(A) $ i $ P(B) $

Mamy 5 próbek w zbiorze danych, z czego:
- 3 próbki należą do klasy $ A $ (w pierwszej, drugiej i czwartej linii),
- 2 próbki należą do klasy $ B $ (w trzeciej i piątej linii).

Obliczamy prawdopodobieństwa priorytetowe dla każdej klasy:

$$
P(A) = \frac{3}{5} = 0.6
$$

$$
P(B) = \frac{2}{5} = 0.4
$$

<br>

## 2. Oblicz prawdopodobieństwa warunkowe $ P(x_1 \mid A) $, $ P(x_1 \mid B) $, $ P(x_2 \mid A) $, $ P(x_2 \mid B) $

### Cecha 1 ($ x_1 $):

$$
P(x_1 = \text{Tak} \mid A) \quad \text{— Ile razy cecha 1 jest równa "Tak" w klasie } A?
$$

W klasie $ A $ mamy 3 próbki, z których 2 mają cechę 1 równą "Tak" (pierwsza i druga próbka). Zatem:

$$
P(x_1 = \text{Tak} \mid A) = \frac{2}{3}
$$

<br>

$$
P(x_1 = \text{Tak} \mid B) \quad \text{— Ile razy cecha 1 jest równa "Tak" w klasie } B?
$$

W klasie $ B $ mamy 2 próbki, z których 1 ma cechę 1 równą "Tak" (trzecia próbka). Zatem:

$$
P(x_1 = \text{Tak} \mid B) = \frac{1}{2}
$$

<br>
<br>

### Cecha 2 ($ x_2 $):

$$
P(x_2 = \text{Tak} \mid A) \quad \text{— Ile razy cecha 2 jest równa "Tak" w klasie } A?
$$

W klasie $ A $ mamy 3 próbki, z których 2 mają cechę 2 równą "Tak" (pierwsza i czwarta próbka). Zatem:

$$
P(x_2 = \text{Tak} \mid A) = \frac{2}{3}
$$

$$
P(x_2 = \text{Tak} \mid B) \quad \text{— Ile razy cecha 2 jest równa "Tak" w klasie } B?
$$

W klasie $ B $ mamy 2 próbki, z których 1 ma cechę 2 równą "Tak" (trzecia próbka). Zatem:

$$
P(x_2 = \text{Tak} \mid B) = \frac{1}{2}
$$

<br>
<br>

## 3. Oblicz $ P(A \mid X) $ i $ P(B \mid X) $

Mamy nowy obiekt, którego cechy to:
- Cecha 1 = "Tak"
- Cecha 2 = "Nie"

Zatem $ P(x_1 = \text{Tak}, x_2 = \text{Nie}) $.

### Obliczanie $ P(A \mid X) $:

Zastosujmy wzór Bayesa dla klasy $ A $:

$$
P(A \mid X) \propto P(A) \cdot P(x_1 = \text{Tak} \mid A) \cdot P(x_2 = \text{Nie} \mid A)
$$

$$
P(x_1 = \text{Tak} \mid A) = \frac{2}{3}, \quad P(x_2 = \text{Nie} \mid A) = 1 - \frac{2}{3} = \frac{1}{3}
$$

$$
P(A \mid X) \propto 0.6 \cdot \frac{2}{3} \cdot \frac{1}{3} = 0.6 \cdot \frac{2}{9} = \frac{1.2}{9} = \frac{2}{15}
$$

### Obliczanie $ P(B \mid X) $:

Zastosujmy wzór Bayesa dla klasy $ B $:

$$
P(B \mid X) \propto P(B) \cdot P(x_1 = \text{Tak} \mid B) \cdot P(x_2 = \text{Nie} \mid B)
$$

$$
P(x_1 = \text{Tak} \mid B) = \frac{1}{2}, \quad P(x_2 = \text{Nie} \mid B) = 1 - \frac{1}{2} = \frac{1}{2}
$$

$$
P(B \mid X) \propto 0.4 \cdot \frac{1}{2} \cdot \frac{1}{2} = 0.4 \cdot \frac{1}{4} = \frac{0.4}{4} = \frac{1}{10}
$$

## 4. Wybór klasy

Porównując wartości $ P(A \mid X) $ i $ P(B \mid X) $, mamy:

$$
P(A \mid X) = \frac{2}{15}, \quad P(B \mid X) = \frac{1}{10} = \frac{1.5}{15}
$$

Ponieważ $ P(B \mid X) > P(A \mid X) $, obiekt zostanie przypisany do klasy $ B $.


### Przykład wykorzystanie klasyfikatora Bayesa w kodzie Python

````python
# Zaimportowanie bibliotek
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Załadowanie zbioru danych Iris
iris = load_iris()
X = iris.data  # cechy: długość i szerokość działki kielicha oraz płatka
y = iris.target  # etykiety gatunków: setosa, versicolor, virginica

# Podział danych na treningowe i testowe
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Tworzenie i trenowanie modelu Naiwnego Bayesa
model = GaussianNB()
model.fit(X_train, y_train)

# Predykcja na danych testowych
y_pred = model.predict(X_test)

# Ocena modelu
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred))
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

````python
data = load_breast_cancer()
X = data.data
y = data.target
feature_names = data.feature_names
target_names = data.target_names
````

````python
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

print(f"Kształt zbioru danych: {df.shape}")
print("-" * 30)
````

````python
print("1. Podstawowe statystyki (EDA):")
print(df.describe().T.head())
````

````python
print("\n2. Liczebność klas (0 = Malignant/Złośliwy, 1 = Benign/Łagodny):")
print(df['target'].value_counts())
````

````python
plt.figure(figsize=(10, 8))
sns.heatmap(df.iloc[:, :10].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Macierz korelacji (dla pierwszych 10 cech)")
plt.show()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
scaler = StandardScaler()
````

````python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

````python
model = GaussianNB()
model.fit(X_train_scaled, y_train)
````

````python
y_pred = model.predict(X_test_scaled)
````

````python
accuracy = accuracy_score(y_test, y_pred)

print("-" * 30)
print(f"Dokładność modelu (Accuracy): {accuracy:.4f}")
print("-" * 30)
print("Raport klasyfikacji:")
print(classification_report(y_test, y_pred, target_names=target_names))

print("Macierz pomyłek (Confusion Matrix):")
# Wyświetlenie macierzy pomyłek
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.ylabel('Prawdziwa klasa')
plt.xlabel('Przewidziana klasa')
plt.title('Macierz pomyłek')
plt.show()
````
