2. Utwórz nowy pandas DataFrame, niech składa się z kolumny składającej się z ciągu liczb od 0 do 12 nazywającej się index, oraz kolumny składającej się z ciągu liczb [-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10, 12, 14] nazywającej się numbers, z czego do uzyskania obu ciągów wykorzystaj funkcję range. Niech kolumna index będzie traktowana jako indeks tego DataFrame

```python
index_values = list(range(13))
numbers_values = list(range(-10, 16, 2))

df = pd.DataFrame({
    "index": index_values,
    "numbers": numbers_values
})

df = df.set_index("index")

df
```

3. Dodaj do tego DataFrame kolumnę zawierającą liczby z kolumny numbers podzielone przez 2 i nazwij ją divided2

```python
df["divided2"] = df["numbers"] / 2

df
```

4. Pomnóż przez siebie wartości z kolumn numbers i divided2, zapisz je w kolumnie multiplied

```python
df["multiplied"] = df["numbers"] * df["divided2"]

df
```

5. Dodaj do DataFrame nową kolumnę nazwaną "Powers of 2", która będzie składała się z wartości uzyskanych z podniesienia do potęgi liczby 2, z czego stopień potęgi będzie określany przez wartości znajdujące się w kolumnie divided2. Zaokrągl te wartości do 2 miejsc po przecinku

```python
df["Powers of 2"] = round(2 ** df["divided2"], 2)

df
```

6. Napisz funkcję sprawdzającą, czy dana liczba jest pierwsza, zracającą True bądź False. Dla liczb nienależących do żadnej z tych kategorii zwróć None. Dodaj do DataFrame kolumnę prime, która będzie zawierała tę flagę obliczoną na podstawie wartości indeksu.

```python
def is_prime(n):

    if n < 2:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True
```

```python
def is_prime(n):

    if n < 2:
        return None

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True
```

7. Odfiltruj z DataFrame te rekordy, które zawierają None w kolumnie prime.

```python
df["prime"] = df.index.map(is_prime)

df
```

8. Dodaj kolejną kolumnę special, która zawierać będzie wartości znajdujące się w kolumnie numbers pomnożone przez 2 jeżeli indeks tego wiersza należy do liczb pierwszych, oraz liczbę 1 w przeciwnym wypadku. Nazwij ją "Do if Prime".

```python
df = df[df["prime"].notna()]

df
```

```python
df["Do if Prime"] = np.where(df["prime"], df["numbers"] * 2, 1)

df
```

11. Wygeneruj ciąg 11 cyfr pierwszych, a następnie dodaj je do DataFrame w kolumnie "Prime Numbers"

```python
def generate_primes(n):

    primes = []
    num = 2

    while len(primes) < n:
        if is_prime(num):
            primes.append(num)
        num += 1

    return primes
```

12. Dodaj kolejną kolumnę "Sum Previous", która będzie sumowała aktualną i poprzednią wartość z kolumny "Prime Numbers". Dodaj kolumnę "Sum Previous Cumulative" liczącą kumulatywną sumę na podstawie wartości z kolumny "Prime Numbers".

```python
df["Prime Numbers"] = generate_primes(len(df))

df
```

13. Oblicz następujące agregacje wartości z kolumny multiplied po grupowaniu wierszy na podstawie parzystości wartości w kolumnie divided2: suma, średnia i liczba (wykorzystaj metodę groupby()).

```python
df["even"] = df["divided2"] % 2 == 0
```

```python
result = df.groupby("even")["multiplied"].agg(["sum", "mean", "count"])

result
```
