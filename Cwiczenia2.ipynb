### NumPy

NumPy jest używane do pracy z tablicami i wykonywania operacji numerycznych.

#### Podstawowe operacje

````python
import numpy as np

# python array
pyarr = [1, 2, 3, 4, 5]
print(pyarr)

# numpy array
arr = np.array([1, 2, 3, 4, 5])
print(arr)

# Podstawowe operacje
print("Mediana:", np.mean(arr))
print("Odchylenie standardowe:", np.std(arr))
print("Suma:", np.sum(arr))
````

#### Wektory i macierze

````python
# Tworzenie tablic NumPy (1D i 2D)
a = np.array([1, 2, 3])            # wektor
b = np.array([[1, 2], [3, 4]])     # macierz 2x2

# Operacje matematyczne
c = a + 10             # dodaj 10 do każdego elementu
d = b * 2              # mnożenie przez skalar
e = a * a              # mnożenie elementów (mnożenie składnikowe)
f = np.dot(a, a)       # iloczyn skalarny

# Operacje na macierzach
dot = np.dot(b, b)       # mnożenie macierzy
transpose = np.transpose(b)    # transpozycja macierzy
inverse = np.linalg.inv([[1, 2], [3, 4]])  # macierz odwrotna (jeśli jest odwracalna)

print("Macierz b * b:\n", dot, "\n")
print("Transpozycja macierzy b:\n", transpose, "\n")
print("Macierz odwrotna macierzy [[1, 2], [3, 4]]:\n", inverse, "\n")
````

#### Slicing

````python
# 1d
arr = np.arange(1, 10)
a = arr[2:4]     # Elementy od indeksu 2 do 4
b = arr[::2]     # Co drugi element
c = arr[-2:]     # Ostatnie 2 elementy
d = arr[::-2]    # Co drugi element w odwrotnej kolejności

print("a) ", a)  # [3 4]
print("b) ", b)  # [1 3 5 7 9]
print("c) ", c)  # [8 9]
print("d) ", d)  # [9 7 5 3 1]

print()

# 2d
arr2d = np.arange(20).reshape(4, 5)  # przekształcenie (wiersze, kolumny)
print("Oryginalna tablica:\n", arr2d)

a2d = arr2d[:3, -3:]  # górny prawy róg 3x3
b2d = arr2d[2]        # 3-ci wiersz

print()
print(a2d)
print(b2d)
````

#### Porównanie wydajności - Pętla for, a operacje na wektorach





````python
import numpy as np
import time

n = 1_000_000
a = np.arange(n)
b = np.arange(n)

# Używanie pętli for
start_loop = time.time()
c_loop = []
for i in range(n):
    c_loop.append(a[i] + b[i])
end_loop = time.time()

# Używanie operacji na wektorach z pomocą NumPy
start_vector = time.time()
c_vector = a + b
end_vector = time.time()

print(f"Czas pętli for: {end_loop - start_loop:.5f} sekundy")
print(f"Czas operacji na wektorach: {end_vector - start_vector:.5f} sekundy")
````

Przydatne parametry

`skiprows=1`:
- To polecenie wskazuje bibliotece Pandas, aby **pominęło pierwszy wiersz** podczas wczytywania pliku CSV.
- Zazwyczaj `skiprows` jest używane, gdy pierwszy wiersz zawiera metadane, dodatkowe informacje lub dane nienależące do tabeli (np. nagłówki w niestandardowej pozycji lub komentarze).
- W takim przypadku pierwszy wiersz nie zostanie uwzględniony w DataFrame. Na przykład, jeśli plik zawiera wiersz wprowadzający lub jakąś opisową treść na górze pliku, ten parametr zapewnia, że będzie to zignorowane podczas wczytywania rzeczywistych danych.

`header=None`:
- Domyślnie `read_csv()` zakłada, że pierwszy wiersz zawiera **nagłówki kolumn** (nazwy kolumn).
- Ustawiając `header=None`, wskazuje to bibliotece Pandas, aby **nie traktowało żadnego wiersza jako nagłówki kolumn**. Zamiast tego Pandas automatycznie przypisze numeryczne etykiety kolumn (0, 1, 2, itd.).
- Jest to przydatne, gdy plik CSV nie zawiera nagłówków lub gdy wiersz nagłówka nie znajduje się w pierwszym wierszu.

#### **1. mediana to punkt srodkowy zbioru danych**

aby obliczyc mediane nalezy:


*   ustawic elementy od najmniejszego do największego
*   jeśli zbiór zawiera nieparzystą liczbę punktów, wybierz element (liczbę) znajdujący się w środku tak uporządkowanego zbioru.
Jeśli zbiór zawiera parzystą liczbę punktów, mediana równa się średniej arytmetycznej dwóch elementów środkowych.


Przykład 1
aby obliczyć medianę zbioru danych, np.
[1,4,2,5,0], należy zacząć od uporządkowania elementów zbioru od najmniejszego do największego:
0,1,2,4,5
Ten zbiór zawiera nieparzystą liczbę elementów, a zatem medianą jest element środkowy
0,1,**2**,4,5, czyli Q2=2.

Przykład 2
aby obliczyć medianę zbioru danych, np.:
10,40,20,50, należy zacząć od uporządkowania elementów zbioru od najmniejszego do największego:
10,20,30,40
W przypadku, gdy zbiór danych składa się z parzystej liczby elementów, mediana jest równa wartości średniej dwóch środkowych elementów.
10, **20, 40**, 50. w tym przypadku mediana wynosi (20+40)/2 = 30

#### **2. średnia arytmetyczna to dzielenie sumy elementów zbioru danych przez liczbę elementów zbioru danych**

Przykładowo, aby obliczyć średnią tego zbioru danych:

####**3. Odchylenie standardowe**
Odchylenie standardowe to miara zróżnicowania danych, pokazująca, o ile przeciętnie wartości w zbiorze odbiegają od średniej arytmetycznej. Im większe odchylenie standardowe, tym większy rozrzut danych i większa zmienność w badanej próbie. Jest to również miara ryzyka i niepewności, powszechnie stosowana w statystyce do oceny rozproszenia wyników.


np. ochylenie standardowe dla liczb [7, 4, -2]
*   najpierw liczymy srednia arytmetyczna
x=9/3=3
*  następnie wariancję
σ2=((7−3)^2+(4−3)^2+(−2−3)^2)/3=(16+1+25)/3=42/3=14
*  odchylenie wynosi σ=√14 ≈3.74

####**4. Kwartyle Q1 i Q3**
Kwartyle to wartości, które dzielą uporządkowany zbiór danych na cztery równe części.
Każda część zawiera 25% wszystkich obserwacji.

Q1 (pierwszy kwartyl) – to wartość, poniżej której znajduje się 25% danych.
Q3 (trzeci kwartyl) – to wartość, poniżej której znajduje się 75% danych.

Kwartyle można obliczyć dla uporządkowanego ciągu pomiarów za pomocą wzoru q_p = x_(n+1)/2, gdzie n to liczba obserwacji.


aby obliczyc kwartyle nalezy uporzadkowac dane rosnąco; znaleźć medianę – podzielić zbiór na połowę ->
* Q1 to mediana dolnej połowy danych (wartości mniejsze niż mediana).

* Q3 to mediana górnej połowy danych (wartości większe niż mediana).

np. Dane:
2, 4, 5, 7, 8, 10, 12, 14
Mediana (środkowa wartość) = (7 + 8) / 2 = 7,5
Dolna połowa: 2, 4, 5, 7 → mediana = (4 + 5) / 2 = 4,5 → Q1 = 4,5
Górna połowa: 8, 10, 12, 14 → mediana = (10 + 12) / 2 = 11 → Q3 = 11

####**5. Rozstęp międzykwartylowy (IQR)**
Rozstęp międzykwartylowy (ang. Interquartile Range) to różnica między trzecim a pierwszym kwartylem.

Innymi słowy wzór na IQR to:
IQR = Q3 – Q1

W ten sposób uzyskujemy 50% naszych obserwacji – tych położonych najbardziej „w środku”. Stąd też im większy IQR, tym większy „środek”, a co za tym idzie większe zróżnicowanie.

np. w zbiorze {1,3,3,5,8,8,8,9}
iqr = 8-3 = 5, ponieważ Q2=(5+8)/2=6,5; (dolna połowa zbioru ={1,3,3,5}) -> q1= (3+3)/2=3; q3=8

####**6. Wariancja**
Wariancja to miara zmienności danych i informuje nas o stopniu, w jakim wartości różnią się od średniej. Im większa wariancja, tym większe są różnice między wartościami. Definiowana jest ona jako średnia arytmetyczna z kwadratów odchyleń poszczególnych wartości cechy od średniej arytmetycznej całej zbiorowości.

Związek z odchyleniem standardowym -> odchylenie standardowe to pierwiastek kwadratowy z wariancji

jest zbiór: 2, 4, 6, 8, 10

a. liczymy średnią -> (2 + 4 + 6 + 8 + 10) / 5 = 6

-4, -2, 0, +2, +4

````python
import numpy as np
x = np.random.randint(1, 101, 100)
x.sort()
print(x)

#mediana
n = len(x) #liczba elementów w zbiorze x
if n % 2 == 0: #jeżeli reszta z dzielenia przez 2 = 0 => przyste
    median = (x[n//2 - 1] + x[n//2]) / 2 #jeżeli n parzyste to średnia 2 środkowych indeksów
else:
    median = x[n//2] #śjeżeli n nieparzyste to mediana to środkowa liczba
print("Mediana:", median)

#średnia
srednia = sum(x) / len(x) #suma przez liczbe elementów
print("Średnia:", srednia)

#q1
dolny = x[:n//2] #od poczatku do srodkowego elementu
d1 = len(dolny)
if d1 % 2 == 0: #jezeli liczba elementow w dolnej polowie jest parzysta
    q1 = (dolny[d1//2 - 1] + dolny[d1//2]) / 2 #srednia 2 srodkowych elementow dolnej polowy
else:
    q1 = dolny[d1//2] #jezeli nieparzysta to srodkowy
print("Q1:", q1)

#q3
gorna = x[n//2:]
m2 = len(gorna)
if m2 % 2 == 0:
    q3 = (gorna[m2//2 - 1] + gorna[m2//2]) / 2
else:
    q3 = gorna[m2//2]
print("Q3:", q3)

# 6. Rozstęp międzykwartylowy (IQR)
iqr = q3 - q1
print("IQR:", iqr)

# 7. Wariancja
wariancja = sum((xi - srednia)**2 for xi in x) / n
wariancja2 = sum((xi - srednia)**2 for xi in x) / (n - 1)

print("Wariancja:", wariancja)
print("Wariancja2:", wariancja2)

# 8. Odchylenie standardowe
odchylenie_std = wariancja ** 0.5
print("Odchylenie standardowe:", odchylenie_std)
odchylenie_std2 = wariancja2 ** 0.5
print("Odchylenie standardowe:", odchylenie_std2)
````

````python
import numpy as np

# Utworzenie tablicy 4x5 z wartościami od 0 do 19
arr = np.arange(20).reshape(4, 5)

print("Oryginalna tablica (4x5):\n", arr)

# Drugi wiersz
second_row = arr[1]

# Trzecia kolumna (0,2)
third_column = arr[:, 2]

# Rogi tablicy (2x2)
top_left_2x2 = arr[:2, :2]
top_right_2x2 = arr[:2, -2:]
bottom_left_2x2 = arr[2:,:2]
bottom_right_2x2 = arr[2:, -2:]

# Wiersze 1 do 3 (czyli 1, 2, 3) i kolumny 2 do 4 (czyli 2, 3, 4)
subarray = arr[0:3, 1:4]

# Wyświetlanie wyników
print("\nDrugi wiersz:")
print(second_row)

print("\nTrzecia kolumna:")
print(third_column)

print("\nLewy górny róg 2x2:")
print(top_left_2x2)

print("\nPrawy górny róg 2x2:")
print(top_right_2x2)
print("\nLewy dolny róg 2x2:")
print(bottom_left_2x2)
print("\nPrawy dolny róg 2x2:")
print(bottom_right_2x2)

print("\nPodtablica z wierszy 1-3 i kolumn 2-4:")
print(subarray)
````

#### Zadanie 8
Wczytaj plik CSV o nazwie `datasaurus.csv` do DataFrame.

Dla zbiorów danych o nazwach `dino` i `circle`:
- oblicz i wyświetl dla wartości **x**:
  - medianę
  - średnią
  - odchylenie standardowe
  - wariancję
- utwórz i wyświetl wykres rozrzutu (scatter plot) dla obu zbiorów danych


````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/data 2/datasaurus.csv")
print(df)

dino = df[df['dataset'] == 'dino']
circle = df[df['dataset'] == 'circle']

print("Zbiór: dino")
print("Mediana:", dino['x'].median())
print("Średnia:", dino['x'].mean())
print("Odchylenie standardowe:", dino['x'].std())
print("Wariancja:", dino['x'].var())

print("\nZbiór: circle")
print("Mediana:", circle['x'].median())
print("Średnia:", circle['x'].mean())
print("Odchylenie standardowe:", circle['x'].std())
print("Wariancja:", circle['x'].var())

plt.figure(figsize=(10, 5))

# Dino
plt.subplot(1, 2, 1)
plt.scatter(dino['x'], dino['y'], color='green')
plt.title('Dino')
plt.xlabel('x')
plt.ylabel('y')

# Circle
plt.subplot(1, 2, 2)
plt.scatter(circle['x'], circle['y'], color='purple')
plt.title('Circle')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
````

#### Zadanie 10
Wczytaj plik CSV o nazwie `anscombes.csv` do DataFrame.

Zwizualizuj wszystkie zbiory danych. Spróbuj ustawić siatkę wykresów (subplot grid) 2x2.


Podpowiedź: `plt.subplots(2, 2, figsize=(10, 8))`

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/data 2/anscombes.csv")

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

datasets = df['dataset'].unique()

for i, dataset in enumerate(datasets):
    ax = axes[i // 2, i % 2]              # wybiera odpowiednią komórkę w siatce 2x2
    subset = df[df['dataset'] == dataset] # filtruje dane dla danego zbioru
    ax.scatter(subset['x'], subset['y'])  # rysuje wykres punktowy
    ax.set_title(f"Zbiór: {dataset}")     # tytuł wykresu
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.tight_layout()
plt.show()
````
