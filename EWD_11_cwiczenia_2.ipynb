```python
import numpy as np
import pandas as pd

X = np.array([20, 19, 5, 12, 13, 13, 3, 17, 10, 1])
Y = np.array([1, 1, 1, 15, 10, 9, 15, 3, 10, 6])
Z = np.array([12, 14, 8, 14, 14, 9, 16, 6, 22, 24])

#Średnia X, Y, Z
sredniaX = sum(w for w in X)/len(X)
print(sredniaX)

sredniaY = sum(w for w in Y)/len(Y)
print(sredniaY)

sredniaZ = sum(w for w in Z)/len(Z)
print(sredniaZ)

#Mediana X, Y, Z
medianaX = np.median(X)
print(medianaX)

medianaY = np.median(Y)
print(medianaY)

medianaZ = np.median(Z)
print(medianaZ)


from collections import Counter

def dominanta(dane):
    licznik = Counter(dane)
    max_wystapien = max(licznik.values())
    return [k for k, v in licznik.items() if v == max_wystapien]

print("Dominanta X:", dominanta(X))
print("Dominanta Y:", dominanta(Y))
print("Dominanta Z:", dominanta(Z))

print("Wariancja X:", np.var(X))
print("Wariancja Y:", np.var(Y))
print("Wariancja Z:", np.var(Z))

print("Odchylenie std X:", np.std(X))
print("Odchylenie std Y:", np.std(Y))
print("Odchylenie std Z:", np.std(Z))
```

```python
import numpy as np

X = np.array([1, 3, 5, 2, 3, 0, 0, 4, 8, 1, 2, 7])
Y = np.array([16, 12, 13, 7, 7, 17, 20, 16, 20, 5, 14, 3])
Z = np.array([25, 2, 19, 30, 22, 25, 14, 2, 16, 30, 12, 21])

# X
print("\n--- X ---")
print("Q1:", np.percentile(X, 25))
print("Mediana:", np.median(X))
print("Q3:", np.percentile(X, 75))
print("Rozstęp:", max(X) - min(X))
print("IQR:", np.percentile(X, 75) - np.percentile(X, 25))

# Y
print("\n--- Y ---")
print("Q1:", np.percentile(Y, 25))
print("Mediana:", np.median(Y))
print("Q3:", np.percentile(Y, 75))
print("Rozstęp:", max(Y) - min(Y))
print("IQR:", np.percentile(Y, 75) - np.percentile(Y, 25))

# Z
print("\n--- Z ---")
print("Q1:", np.percentile(Z, 25))
print("Mediana:", np.median(Z))
print("Q3:", np.percentile(Z, 75))
print("Rozstęp:", max(Z) - min(Z))
print("IQR:", np.percentile(Z, 75) - np.percentile(Z, 25))
```

```python
import numpy as np

X = np.array([3, 0, 4, 5, 5, 1, 2, 2])
Y = np.array([2, 4, 5, -1, 1, 1, 7, 7])

# X
print("\n--- X ---")
norm_X = (X - min(X)) / (max(X) - min(X))
stand_X = (X - np.mean(X)) / np.std(X)

print("Normalizacja:", norm_X)
print("Standaryzacja:", stand_X)

# Y
print("\n--- Y ---")
norm_Y = (Y - min(Y)) / (max(Y) - min(Y))
stand_Y = (Y - np.mean(Y)) / np.std(Y)

print("Normalizacja:", norm_Y)
print("Standaryzacja:", stand_Y)
```

```python
import numpy as np
import itertools

omega = list(itertools.product(range(1,7), repeat=2))
print("Liczba zdarzeń:", len(omega))

# b) prawdopodobieństwa
# i) przynajmniej jedna parzysta
A = [o for o in omega if o[0]%2==0 or o[1]%2==0]
print("P(parzysta >=1):", len(A)/36)

# ii) obie nieparzyste
B = [o for o in omega if o[0]%2==1 and o[1]%2==1]
print("P(obie nieparzyste):", len(B)/36)

# iii) suma = 7
C = [o for o in omega if o[0]+o[1]==7]
print("P(suma=7):", len(C)/36)


# c)
# i)
warunek = [o for o in omega if o[0]%2==0]
zdarzenie = [o for o in warunek if o[1]%2==1]
print("P(druga nieparzysta | pierwsza parzysta):", len(zdarzenie)/len(warunek))

# ii)
warunek2 = [o for o in omega if o[0]==1]
zdarzenie2 = [o for o in warunek2 if o[1]==6]
print("P(druga=6 | pierwsza=1):", len(zdarzenie2)/len(warunek2))


# d)
kostka = np.array([1,2,3,4,5,6])
E = np.mean(kostka)

Var1 = np.var(kostka)
Var2 = np.mean((kostka - E)**2)

print("E(X):", E)
print("Var(X) - wzór 1:", Var1)
print("Var(X) - wzór 2:", Var2)
```
