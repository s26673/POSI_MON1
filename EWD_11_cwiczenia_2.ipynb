```python
import numpy as np
import pandas as pd

X = np.array([20, 19, 5, 12, 13, 13, 3, 17, 10, 1])
Y = np.array([1, 1, 1, 15, 10, 9, 15, 3, 10, 6])
Z = np.array([12, 14, 8, 14, 14, 9, 16, 6, 22, 24])

#Średnia X, Y, Z
sredniaX = sum(w for w in X)/len(X)
print(sredniaX)

sredniaY = sum(w for w in Y)/len(Y)
print(sredniaY)

sredniaZ = sum(w for w in Z)/len(Z)
print(sredniaZ)

#Mediana X, Y, Z
medianaX = np.median(X)
print(medianaX)

medianaY = np.median(Y)
print(medianaY)

medianaZ = np.median(Z)
print(medianaZ)

ppp
```
