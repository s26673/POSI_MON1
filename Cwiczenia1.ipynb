### Zadanie 8
Napisz funkcję `bmi(weight, height)`, która:
* oblicza wskaźnik BMI wg wzoru: BMI = $\frac{\text{waga}}{(\text{wzrost})^2}$
* zwraca komunikat:
  * `< 18.5` → "niedowaga"
  * `18.5-24.9` → "waga prawidłowa"
  * `25-29.9` → "nadwaga"


````python
def bmi(weight, height):
  bmi = float(weight) / (float(height) ** 2)
  if bmi < 18.5:
    return ("Niedowaga")
  elif bmi >= 18.5 and bmi <= 24.9:
    return ("Waga prawidłowa")
  elif bmi >= 25 and bmi <= 29.9:
    return ("Nadwaga")

print(bmi(60, 1.76))

weight = input("Podaj swoją wagę w kg: ")
height = input("Podaj swój wzrost w m (np.1.76 dla 176cm): ")
print(bmi(weight, height))
````

### Zadanie 10 - Odwracanie listy
 Napisz funkcję, która **odwróci listę** bez używania metody `reverse()` ani ` :-1`.

 (np. `["a", "b", "c", "d"] → ["d", "c", "b", "a"]`)

````python
def custom_reverse(my_list):
  for i in range(int(len(my_list)/2)):
    tmp = my_list[i]
    my_list[i] = my_list[len(my_list) - 1 - i]
    my_list[len(my_list) - 1 - i] = tmp

my_list = ["a", "b", "c", "d"]
custom_reverse(my_list)
print(my_list)


#reverse()
my_list = ["a", "b", "c", "d"]
my_list.reverse()
print(my_list)

#:-1
my_list = ["a", "b", "c", "d"]
print(my_list[::-1])
````

### Zadanie 11 - Zliczanie wystąpień
Dla listy:

``` fruits = ["apple", "banana", "apple", "orange", "banana", "apple"] ```


Policz, ile razy występuje każdy owoc i zapisz wynik w słowniku, np.:

```  {"apple": 3, "banana": 2, "orange": 1} ```

````python
def count(items):
  counts = {}
  for x in items:
    if counts.get(x) == None:
      counts[x] = 1
    else:
      counts[x] +=1

  return counts

fruits = ["apple", "banana", "apple", "orange", "banana", "apple"]
print(count(fruits))


#.get
def count(items):
    counts = {}
    for x in items:
        counts[x] = counts.get(x, 0) + 1
    return counts

fruits = ["apple", "banana", "apple", "orange", "banana", "apple"]
print(count(fruits))
````

### Zadanie 12 - Normalizacja danych
Masz listę wartości:
```  data = [10, 20, 30, 40, 50] ```

Napisz funkcję `normalize(data)`, która zwróci listę wartości przeskalowanych do przedziału `[0, 1]` według wzoru:

$ x' = \frac{x - min}{max - min}$

````python
data = [10, 20, 30, 40, 50]

def normalize(data):
    min_val = data[0]
    max_val = data[0]

    for x in data:
        if x < min_val:
            min_val = x
        if x > max_val:
            max_val = x

    if max_val == min_val:
        return [0.0 for _ in data]

    normalized = []
    for x in data:
        x_norm = (x - min_val) / (max_val - min_val)
        normalized.append(x_norm)

    return normalized

print(normalize(data))
````

### Zadanie 13 - Bag of Words (liczenie słów)
Masz listę zdań:
```
sentences = [
 "I like Python",
 "Python is great for AI",
 "I like AI"
]
 ```

Napisz funkcję `bag_of_words(sentences)`, która zwróci słownik z częstością występowania każdego słowa.

````python
def count(items):
    counts = {}
    for x in items:
        if counts.get(x) == None:
            counts[x] = 1
        else:
            counts[x] += 1
    return counts

sentences = [
    "I like Python",
    "python is great for AI",
    "I like AI"
]

def bag_of_words(sentences):
    words = []
    for sentence in sentences:
        words.extend(sentence.lower().split()) #lower(), żeby Python i python było liczone jako jedno słowo
    return count(words)

print(bag_of_words(sentences))

#2 sposób
def bag_of_words(sentences):
    counts = {}
    for sentence in sentences:
        words = sentence.split()
        for word in words:
            counts[word] = counts.get(word, 0) + 1
    return counts

print(bag_of_words(sentences))

zdanie = input("Podaj zdanie: ")
print(bag_of_words([zdanie]))
````

### Zadanie 14 - Liczby doskonałe
Napisz funkcję `is_perfect(n)`, która sprawdzi, czy liczba jest **doskonała**, czyli równa sumie swoich dzielników właściwych.

Przykład: `6` → dzielniki 1+2+3 = 6 → liczba doskonała.

````python
def is_perfect(n):
  sum = 0
  for i in range(1, n):
    if n % i == 0:
      sum += i
  if sum == n:
    return True
  else:
    return False

print(is_perfect(6))
print(is_perfect(7))
print(is_perfect(28))

n = int(input("Podaj liczbę: "))
print("Czy liczba jest doskonała?", is_perfect(n))
````
